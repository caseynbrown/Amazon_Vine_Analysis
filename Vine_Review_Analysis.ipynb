```python
vine_data_to_load = "C:/Users/casey/OneDrive/Desktop/Course Work/Module 17/Starter_Code (3)/vine_table.csv"
vine_data_df = pd.read_csv(vine_data_to_load)
```

```python
vine_data_df.head()
```

```python
new_df=vine_data_df[vine_data_df['total_votes'] > 19]
new_df.head(10)
```

```python
new_df['results'] = new_df['helpful_votes'] / new_df['total_votes']
new_df.head(10)
```

```python
next_new_df = new_df[new_df['results'] > 0.5]
next_new_df.head(10)
```

```python
paid_df = next_new_df[next_new_df['vine'] == 'Y']
paid_df.head(10)
```

```python
unpaid_df = next_new_df[next_new_df['vine'] == 'N']
unpaid_df.head(10)
```

```python
# count of paid Vine votes
paid_df['total_votes'].count()
```

```python
# count of unpaid Vine votes
unpaid_df['total_votes'].count()
```

```python
#number of paid Vine 5-star reviews
len(paid_df[paid_df['star_rating']==5])
```

```python
#number of unpaid Vine 5-star reviews
len(unpaid_df[unpaid_df['star_rating']==5])
```

```python
#Percent 5 star review paid Vine
(len(paid_df[paid_df['star_rating']==5]))/(paid_df['total_votes'].count())
```
